```python
import pandas as pd

# List of file paths
file_paths = [
    '/content/USDA_Soybean_County_2016.csv',
    '/content/USDA_Soybean_County_2017.csv',
    '/content/USDA_Soybean_County_2018.csv',
    '/content/USDA_Soybean_County_2019.csv',
    '/content/USDA_Soybean_County_2020.csv',
    '/content/USDA_Soybean_County_2021.csv',
    '/content/USDA_Soybean_County_2022.csv'
]

# List to hold DataFrames
dataframes = []

# Loop through file paths and read each file into a DataFrame
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Concatenate all DataFrames into one
merged_data = pd.concat(dataframes, ignore_index=True)

# Display the merged data
print(merged_data.head())
```

```python
state_year_yield = merged_data.groupby(['state_ansi', 'year'])['YIELD, MEASURED IN BU / ACRE'].mean().reset_index()

# Pivot the data for easier plotting
pivot_data = state_year_yield.pivot(index='year', columns='state_ansi', values='YIELD, MEASURED IN BU / ACRE')

# Plotting the data
pivot_data.plot(figsize=(14, 7), marker='o')
plt.title('Variation of Soybean Yield Across States (2017-2022)')
plt.xlabel('Year')
plt.ylabel('Average Yield')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(pivot_data, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of Soybean Yield Across States (2017-2022)')
plt.xlabel('state')
plt.ylabel('year')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Load your yield data
# Example structure: yield_data = pd.DataFrame({'State': ['Alabama', 'Alaska', ...], 'Yield': [123, 456, ...]})
yield_data = dataframe.groupby('state_name')['YIELD, MEASURED IN BU / ACRE'].mean().reset_index()

# To match state names with plotly's recognized state names, use state abbreviations
state_abbreviations = {
    'ALABAMA': 'AL', 'ARKANSAS': 'AR','DELAWARE': 'DE','GEORGIA': 'GA','ILLINOIS': 'IL', 'INDIANA': 'IN', 'IOWA': 'IA', 'KANSAS': 'KS', 'KENTUCKY': 'KY', 'LOUISIANA': 'LA',
    'MARYLAND': 'MD', 'MICHIGAN': 'MI', 'MINNESOTA': 'MN', 'MISSISSIPPI': 'MS',
    'MISSOURI': 'MO', 'NEBRASKA': 'NE', 'NEW JERSEY': 'NJ','NEW YORK': 'NY', 'NORTH CAROLINA': 'NC', 'NORTH DAKOTA': 'ND', 'OHIO': 'OH',
    'OKLAHOMA': 'OK', 'PENNSYLVANIA': 'PA','SOUTH CAROLINA': 'SC',
    'SOUTH DAKOTA': 'SD', 'TENNESSEE': 'TN', 'TEXAS': 'TX','VIRGINIA': 'VA','WISCONSIN': 'WI'
}

# Add state abbreviations to the data
yield_data['State Abbrev'] = yield_data['state_name'].map(state_abbreviations)

# Plotting the choropleth map
fig = px.choropleth(
    yield_data,
    locations='State Abbrev',
    locationmode='USA-states',
    color='YIELD, MEASURED IN BU / ACRE',
    color_continuous_scale='YlGnBu',
    scope='usa',
    labels={'YIELD, MEASURED IN BU / ACRE': 'Average Yield'}
)

fig.update_layout(
    title_text='Average Soybean Yield by State in the USA',
    geo=dict(
        showcoastlines=True,
    )
)

fig.show()
```

```python
import plotly.express as px
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Load your data
# Example structure: data = pd.DataFrame({'state_name': ['ALABAMA', 'GEORGIA', ...], 'year': [2016, 2016, ...], 'YIELD, MEASURED IN BU / ACRE': [123, 456, ...]})
# Replace with your data loading method

# Group by state and year, then calculate the average yield
yield_data = mer_data.groupby(['state_name', 'year'])['YIELD, MEASURED IN BU / ACRE'].mean().reset_index()

# State abbreviations mapping
state_abbreviations = {
    'ALABAMA': 'AL', 'ARKANSAS': 'AR', 'DELAWARE': 'DE', 'GEORGIA': 'GA', 'ILLINOIS': 'IL', 'INDIANA': 'IN', 'IOWA': 'IA', 'KANSAS': 'KS', 'KENTUCKY': 'KY', 'LOUISIANA': 'LA',
    'MARYLAND': 'MD', 'MICHIGAN': 'MI', 'MINNESOTA': 'MN', 'MISSISSIPPI': 'MS', 'MISSOURI': 'MO', 'NEBRASKA': 'NE', 'NEW JERSEY': 'NJ', 'NEW YORK': 'NY', 'NORTH CAROLINA': 'NC', 'NORTH DAKOTA': 'ND', 'OHIO': 'OH',
    'OKLAHOMA': 'OK', 'PENNSYLVANIA': 'PA', 'SOUTH CAROLINA': 'SC', 'SOUTH DAKOTA': 'SD', 'TENNESSEE': 'TN', 'TEXAS': 'TX', 'VIRGINIA': 'VA', 'WISCONSIN': 'WI'
}

# Map state names to abbreviations
yield_data['State Abbrev'] = yield_data['state_name'].map(state_abbreviations)

# Create subplots for each year
years = sorted(yield_data['year'].unique())
fig = make_subplots(rows=1, cols=len(years), subplot_titles=[f'Year {year}' for year in years], specs=[[{'type': 'choropleth'}] * len(years)])

# Add a choropleth map for each year
for i, year in enumerate(years):
    year_data = yield_data[yield_data['year'] == year]
    choropleth = px.choropleth(
        year_data,
        locations='State Abbrev',
        locationmode='USA-states',
        color='YIELD, MEASURED IN BU / ACRE',
        color_continuous_scale='YlGnBu',
        scope='usa',
        labels={'YIELD, MEASURED IN BU / ACRE': 'Average Yield'}
    )

    # Add traces to subplot
    for trace in choropleth.data:
        fig.add_trace(trace, row=1, col=i + 1)

# Update layout
fig.update_layout(
    title_text='Average Soybean Yield by State in the USA (2016-2022)',
    geo=dict(showcoastlines=True),
    showlegend=False,
    height=600,
    width=1500
)

fig.show()
```
